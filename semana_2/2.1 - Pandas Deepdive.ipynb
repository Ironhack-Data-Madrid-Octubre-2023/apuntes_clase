### AplicaciÃ³n de funciones

````python
df.head()
````

````python
def restar(x):
    """
    Entra un nÂº y sale un nÂº
    
    """
    
    return x-2 
````

````python
%%time

df.quantity = [restar(x) for x in df.quantity]

df.head()
````

````python
%%time

df.quantity = list(map(restar, df.quantity))

df.head()
````

````python
%%time

df.quantity = map(restar, df.quantity)

df.head()
````

````python
%%time

df.quantity = pd.Series(map(restar, df.quantity))

df.head()
````

````python
%%time

df.quantity = df.quantity.apply(restar) # esto es un map, el iterable es la columna

df.head()
````

````python
%%time

df.quantity = df.quantity.apply(lambda x: x-2) # esto es un map, el iterable es la columna

df.head()
````

````python
%%time

df.quantity = df.quantity - 2

df.head()
````

````python
%%time

df['nueva'] = df.quantity/df.unitprice

df.head()
````

````python
%%time

df['nueva2'] = df.quantity - df.unitprice

df.head()
````
